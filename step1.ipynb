```python
# from alpha_vantage.timeseries import TimeSeries
from confluent_kafka import Producer
import json
import requests
import config
```

```python
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=AAPL&interval=5min&apikey={config.api_key}'
r = requests.get(url)
data = r.json()
```

```python
data1 = r.json()
```

```python

new = []
# print(data1)
newdict = {}
import pandas as pd
for k in data1["Time Series (5min)"].keys():
    new.append((data1["Meta Data"]["2. Symbol"],k, data1["Time Series (5min)"][k]["4. close"]))

df = pd.DataFrame(new)
df.columns = ['Symbol', 'Timeseries', 'Close']
#print(df.head())

print("========================")
newdict["symbol"] = df["Symbol"][0]
newdict["timeseries"] = df['Timeseries'].to_list()
newdict["close"] = df['Close'].to_list()

print(newdict)


```

```python

#print(new)
# Produce data to Kafka
producer = Producer(config.kafka_config)
producer.produce(config.kafka_topic, key="AAPL", value=json.dumps(newdict))
producer.flush()
```
